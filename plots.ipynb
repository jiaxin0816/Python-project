### 2) How are the different variables (gender, marital status, etc.) distributed?

```python
fig, ax = plt.subplots(figsize=[7,5])

sns.histplot(
    data = churn,
    x = "Gender",
    binwidth = 1,
    color = "steelblue",
    edgecolor = "black")

ax.set(
    title = "Gender distribution",
    xlabel = "Gender",
    ylabel = "Frequency"
)
```

```python
fig, ax = plt.subplots(figsize=[7,5])

sns.kdeplot(
    data=churn,
    x="MonthlyIncome",
    color = "steelblue"
)

ax.set(
    title="Monthly Income Distribution",
    xlabel = "Monthly Income"
)
```

```python
fig, ax = plt.subplots(figsize=[8,5])

sns.kdeplot(
    data=churn,
    x="MonthlyIncome",
    hue = "EducationField",
    palette = "Set1",
    fill = True
)

ax.set(
    title="Monthly Income distribution by Education Field",
    xlabel = "Monthly Income"
)

ax.set_xlim(left=0)
```

```python
fig, ax = plt.subplots(figsize=[10,8])

sns.boxplot(
    data=churn,
    x="EducationField",
    y = "MonthlyIncome",
    width = .3
)

ax.set(
    title="Monthly Income distribution by Education Field",
    xlabel = "EducationField",
    ylabel = "Monthly Income"
)

```

We can observe the following:
- Employees with Marketing degrees have the higher median salary (around 6000?)
- Medical and Life sciences show the most outliers, indicating that some individuals in these fields have exceptionally higher incomes than their peers. This is not so suprising since advanced levels of expertise in these fields often are very appreciated in businesses because of their technical capacities?
- The overall income range of employees in the human ressources field is significanlty smaller compared to other fields (maximum of around 10000 (without outliers)) 

```python
fig, ax = plt.subplots(figsize=[25,10])

sns.boxplot(
    data=churn,
    x="JobRole",
    y = "MonthlyIncome",
    width = .3
)

ax.set(
    title="Monthly Income distribution by Job Role",
    xlabel = "Job Role",
    ylabel = "Monthly Income"
)

```

```python
fig, ax = plt.subplots(figsize=[25,10])

sns.boxplot(
    data=churn,
    x="JobRole",
    y = "Age",
    width = .3
)

ax.set(
    title="Age distribution by Job Role",
    xlabel = "Job Role",
    ylabel = "Age"
)

```

- Employees in Manager positions have the highest median salary (around 17500), followed by research directors (over 15000). These 2 roles subsequently hold the highest median age of employees, which is logical, because high salary comes with experience, and experience is acquired with age. 
- Sales executive, Manufacturing directors and Healthcare Representatives have a similar median salary of around 6000-6250, which can go up to 14000-14250.
- Finally, Research Scientists, Laboratory Technicians, Sales representatives and Human Ressources have the lowest median salary, around 3000. It is worth noting however, that employees in human ressources have the biggest salary range among these 4, which can reach over 10000. Human ressources is an important and necessary department at every enterprise, so it is natural to find that HR staff can have a range of salaries, which means possibilities of promotions and career development. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'churn'
# Filter the dataset where 'churn' is 'YES'
churn_yes = churn[churn['Churn'] == 'Yes']

# Count how many employees left from each department
churn_count_by_dept = churn_yes['Department'].value_counts()

# Create a bar plot
plt.figure(figsize=[10, 6])
sns.barplot(x=churn_count_by_dept.index, y=churn_count_by_dept.values, palette='Blues_d')

# Add labels and title
plt.title('Number of Employees Who Left by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees Who Left', fontsize=12)



# Show the plot
plt.show()
```

#### Employee satisfaction by department

```python
fig, ax= plt.subplots(figsize=(8,5))

sns.countplot(
    data=churn,
    x='JobSatisfaction'
)

ax.set(
    title="Employee Job satisfaction count",
    xlabel='Satisfaction level',
    ylabel = 'count'
)
sns.despine()
```

```python


# Assuming your DataFrame is named 'churn' and has columns 'department' and 'satisfaction_level'
# You may need to replace 'satisfaction_level' with the actual column name in your dataset.

# Create a pivot table to count satisfaction levels by department
satisfaction_by_dept = churn.pivot_table(
    index='Department', 
    columns='RelationshipSatisfaction', 
    aggfunc='size', 
    fill_value=0
)

# Create the stacked bar plot
satisfaction_by_dept.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='viridis')

# Add labels and title
plt.title('Employee Satisfaction by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.legend(title='Satisfaction Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
gender_by_dept = churn.pivot_table(
    index='Department', 
    columns='Gender', 
    aggfunc='size', 
    fill_value=0
)

gender_by_dept.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='viridis')

plt.title('Employee gender by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

plt.xticks(rotation=45)

plt.legend(title='Gender by department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
